```python
# How many 'Action' movies with rating higher than 7.5
```

```python
df1 = (movies['genres'].str.split('|').apply(lambda x:'Action' in x))
df2 = (movies['imdb_rating'] > 7.5)

movies[df1 & df2].shape[0]
```
